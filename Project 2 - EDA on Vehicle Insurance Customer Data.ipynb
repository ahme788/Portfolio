####  For example: for a column X calculate Q1 = 25th percentile and Q3 = 75th percentile then IQR = Q3 – Q1 ) then to check outlier, #### anything lower than a Q1 – 1.5IQR or greater than Q3 + 1.5 IQR would be an outlier

```python
def find_outliers_IQR(customer_details):

    q1=customer_details.quantile(0.25)

    q3=customer_details.quantile(0.75)

    IQR=q3-q1

    outliers = customer_details[((customer_details<(q1-1.5*IQR)) | (customer_details>(q3+1.5*IQR)))]

    return outliers
```

```python
outliers =  find_outliers_IQR(customer_details['age'])
                             
outliers = find_outliers_IQR(customer_details['Driving Licence Present'])
                             
outliers = find_outliers_IQR(customer_details['Region Code'])
                             
outliers = find_outliers_IQR(customer_details['Vehicle age'])
```
